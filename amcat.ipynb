```python
# Step 3: Univariate Analysis

# 3.1 Visualizing numerical columns distributions
numerical_columns = ['Salary', '10percentage', '12percentage', 'collegeGPA', 'English',
                     'Logical', 'Quant', 'ComputerProgramming', 'ComputerScience']
```

```python
for col in numerical_columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[col], kde=True)
    plt.title(f'Histogram of {col}')
    plt.show()
```

```python
# Step 4: Bivariate Analysis

# 4.1 Scatterplots for numerical vs numerical
for num_col in numerical_columns:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=df[num_col], y=df['Salary'])
    plt.title(f'Salary vs {num_col}')
    plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt

# Initialize the scaler
scaler = StandardScaler()

# Fit and transform the numerical columns (except 'Salary', as you might want to keep it unscaled)
scaled_columns = numerical_columns[1:]  # Ignore 'Salary' for now
df_scaled = df.copy()
df_scaled[scaled_columns] = scaler.fit_transform(df[scaled_columns])
```

```python
# 4.2 Boxplots for categorical vs numerical
for cat_col in categorical_columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df[cat_col], y=df['Salary'])
    plt.xticks(rotation=90)
    plt.title(f'Salary vs {cat_col}')
    plt.show()
```

```python

# 5.2 Relationship between Gender and Specialization using Chi-Square Test
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df['Gender'], df['Specialization'])

# Perform Chi-Square test
chi2, p, dof, ex = chi2_contingency(contingency_table)

print(f"Chi-Square Test: p-value = {p}")
if p < 0.05:
    print("There is a significant relationship between Gender and Specialization.")
else:
    print("There is no significant relationship between Gender and Specialization.")
```
